```python
# Function to count pairs in a list that sum up to a target value
def count_pairs_with_sum(target_sum, number_list):
    visited_elements = set()       # To track elements we've already seen
    result_pairs = []              # To store valid pairs

    for number in number_list:
        complement = target_sum - number  # Calculate the required pair value
        if complement in visited_elements:
            result_pairs.append((complement, number))  # Add valid pair
            visited_elements.remove(complement)        # Remove to avoid reuse
        else:
            visited_elements.add(number)               # Store current number

    return len(result_pairs), result_pairs             # Return count and pairs


number_list = []  # List to store user input
n = int(input("Enter the size of the list "))  # Get size of the list

print("Enter the list elements")
for _ in range(n):
    value = int(input())       # Read each list element
    number_list.append(value)  # Append to the list

target_sum = int(input("Enter the target sum: "))  # Get the target sum

# Call the function to find the pairs
pair_count, pair_list = count_pairs_with_sum(target_sum, number_list)

# Display results
print("Pairs that sum to", target_sum, ":", pair_list)
print("Number of such pairs:", pair_count)
```

```python
# Function to take float inputs from the user
def take_input():
    number_list = []
    n = int(input("Enter how many numbers you want to input: "))  # Ask user for input size
    for _ in range(n):
        value = float(input())        # Read each float number
        number_list.append(value)     # Append to the list
    return number_list

# Function to calculate range (max - min) if list has at least 3 numbers
def calculate_range(number_list):
    if len(number_list) < 3:
        return "Range determination not possible"  # Error if not enough data
    return max(number_list), min(number_list)      # Return max and min values

# Get the list of numbers from user
numbers = take_input()

# Get the result (either an error message or a (max, min) tuple)
result = calculate_range(numbers)

# Print appropriate result
if isinstance(result, str):   # If result is a string, it's an error message
    print(result)
else:
    max_val, min_val = result
    print(f"{min_val} - {max_val}")      # Show the range formula
    print(max_val - min_val)             # Show the actual range
```

```python
# Function to take input for a square matrix from the user
def take_matrix_input():
    size = int(input("Enter size of square matrix: "))
    print("Enter the elements of the matrix")
    matrix = []
    for i in range(size):
        row = []
        for j in range(size):
            value = int(input())
            row.append(value)
        matrix.append(row)
    return matrix

# Function to multiply two square matrices A and B
def multiply_matrices(A, B):
    size = len(A)
    result = []

    for i in range(size):
        row = []
        for j in range(size):
            sum_val = 0
            # Multiply the i-th row of A with the j-th column of B
            for k in range(size):
                sum_val += A[i][k] * B[k][j]
            row.append(sum_val)
        result.append(row)

    return result

# Function to compute matrix raised to the given power
def matrix_power(matrix, power):
    result = matrix
    for _ in range(power - 1):
        result = multiply_matrices(result, matrix)
    return result

# Take matrix and power input from the user
matrix = take_matrix_input()
m = int(input("Enter the power to raise the matrix: "))

# Calculate matrix raised to power m
result = matrix_power(matrix, m)

# Print the resulting matrix
for row in result:
    print(row)
```

```python
# Function to find the highest occurring alphabet character in the string
def get_highest_occurring_char(input_string):
    frequency = {}  # Dictionary to store character frequencies

    # Loop through each character in the string
    for char in input_string:
        if char.isalpha():  # Consider only alphabetic characters
            if char in frequency:
                frequency[char] += 1  # Increment count if already exists
            else:
                frequency[char] = 1   # Initialize count

    max_char = ''     # To store the character with the highest frequency
    max_count = 0     # To store its count

    # Loop through the frequency dictionary to find the max
    for char in frequency:
        if frequency[char] > max_count:
            max_char = char
            max_count = frequency[char]

    return max_char, max_count  # Return most frequent character and its count

# Take input string from the user
user_input = input("Enter the string - ")

# Call the function to get the most frequent character and its count
char, count = get_highest_occurring_char(user_input)

# Print the results
print(char)
print(count)
```

```python
# Function to take 25 integer inputs from the user
def take_input():
    print("Enter the elements of the list")
    numbers = []
    for _ in range(25):
        num = int(input())       # Read one number at a time
        numbers.append(num)      # Add to the list
    return numbers

# Function to calculate the mean (average) of the numbers
def calculate_mean(numbers):
    total = 0
    for num in numbers:
        total += num             # Sum all numbers
    return total / len(numbers)  # Divide by count

# Function to calculate the median of the numbers
def calculate_median(numbers):
    numbers.sort()               # Sort the list first
    mid = len(numbers) // 2
    if len(numbers) % 2 == 0:    # If even number of elements, take average of two middle ones
        return (numbers[mid - 1] + numbers[mid]) / 2
    else:                        # If odd, take the middle element
        return numbers[mid]

# Function to calculate the mode (most frequent number)
def calculate_mode(numbers):
    frequency = {}               # Dictionary to count frequencies
    for num in numbers:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1

    max_count = 0
    mode_value = None
    for num in frequency:
        if frequency[num] > max_count:  # Update mode if higher frequency is found
            max_count = frequency[num]
            mode_value = num

    return mode_value

# Main logic
data = take_input()

# Print mean, median, and mode
print("Mean =", calculate_mean(data))
print("Median =", calculate_median(data))
print("Mode =", calculate_mode(data))
```
